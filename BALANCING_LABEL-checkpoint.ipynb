```python
CSV_FILE = r"E:\Sujitha_MISS_DATASET\2April_FULL_DATASET\siemens_filtered_df.csv"
OUTPUT_CSV = r"E:\Sujitha_MISS_DATASET\2April_FULL_DATASET\siemens_filtered_df_BALANCED.csv"
```

```python
import pandas as pd
# Ensure Pandas displays the full string
pd.set_option("display.max_colwidth", None)

# Step 1: Read CSV
df = pd.read_csv(CSV_FILE, parse_dates=["Archive Date", "Study Date"])  # Ensure date is correctly parsed
df.head()
```

```python
# Step 2: Convert Subject IDs to string for consistency
df["Subject ID"] = df["Subject ID"].astype(str).str.strip()
df[["Subject ID", "Research Group"]].head()
```

```python
# Step 3: Sort by Subject and Acquisition Date (latest first)
df_sorted = df.sort_values(by=["Subject ID", "Study Date"], ascending=[True, False])
df_sorted.head()
```

```python
# Step 4: Separate Control and PD groups
df_control = df_sorted[df_sorted["Research Group"] == "Control"]
df_pd = df_sorted[df_sorted["Research Group"] == "PD"]

print(f"Total Control Subjects: {len(df_control)}")
print(f"Total PD Subjects Before Removing Duplicates: {len(df_pd)}\n")
```

```python
# Step 5: Keep the latest scan for each PD subject
df_pd_unique = df_pd.drop_duplicates(subset=["Subject ID"], keep="first")
df_pd_unique.head()
```

```python
print(f"Total PD Subjects After Removing Duplicates: {len(df_pd_unique)}\n")
```

```python

# Step 6: Downsample PD to balance the dataset
pd_sample_size = min(len(df_pd_unique), len(df_control))  # Adjust the sample size as needed
df_pd_sampled = df_pd_unique.sample(n=pd_sample_size, random_state=42)

print(f"Total PD Subjects After Sampling: {len(df_pd_sampled)}\n")
```

```python
# Step 7: Merge PD and full Control dataset
df_balanced = pd.concat([df_control, df_pd_sampled])

df_balanced.head()
```

```python
df_balanced.to_csv(OUTPUT_CSV, index=False)
print(f"Balanced dataset saved as {OUTPUT_CSV}")
```
