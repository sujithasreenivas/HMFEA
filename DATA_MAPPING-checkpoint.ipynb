## 1. Extract the Datapoints Folder
This code extracts the folder_path to each data points containing the dcm file of single scan

```python
import os
from tqdm import tqdm

def find_image_folders(root_dir, image_exts=(".nii.gz")):
    image_folders = []

    for dirpath, _, filenames in tqdm(os.walk(root_dir), desc="Walking Through Directories"):
        if any(f.lower().endswith(image_exts) for f in filenames):  # Check if folder contains images
            image_folders.append(dirpath)

    return image_folders

def count_files_in_folder(folder_path):
    try:
        return len([f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))])
    except FileNotFoundError:
        print("Folder not found!")
        return 0
```

```python

root_directory = DATASET_ROOTDIR
image_folders_list = find_image_folders(root_directory)
print('Total Number of Data Points:', len(image_folders_list))

image_folders_list[:5]
```

## 2. Mapping the Datapoints with Class
This code maps the folder_path (i.e. single data point) to its corresponding class (PD or Control) by using the CSV file provided

```python
import pandas as pd

# Read CSV and create a mapping from Image Data ID to Research Group
df = pd.read_csv(CSV_FILE)

# Convert Image Data ID column to string for consistent lookup
df["Image ID"] = df["Image ID"].astype(str).str.strip()  # Remove spaces
image_id_to_group = df.set_index("Image ID")["Research Group"].to_dict()

# Debug: Print first 10 Image Data IDs from the CSV
print("Image Data IDs from CSV:", list(image_id_to_group.keys())[:10])  # Print first 10 IDs

filtered_data = []
for path in image_folders_list:
    parts = path.split("\\")  # Ensure Windows path splitting

    if len(parts) >= 1:  # Ensure path is valid
        image_id = parts[-1].strip()  # Extract Image Data ID from the last part
        
        if image_id in image_id_to_group:  
            filtered_data.append({"Path": path, "Research Group": image_id_to_group[image_id]})
        else:
            print(f"Image Data ID {image_id} not found in CSV!")

# Create DataFrame
mapped_df = pd.DataFrame(filtered_data)

# Debug: Print DataFrame size
print(f"Total Mapped Rows: {len(mapped_df)}")
```
