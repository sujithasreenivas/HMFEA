```python
from pathlib import Path

# Define important variables
ROOT_DIR = Path(r"E:\Sujitha_MISS_DATASET\2April_FULL_DATASET\PPMI_Full_NIfTI")
CSV_FILE = r"E:\Sujitha_MISS_DATASET\2April_FULL_DATASET\siemens_filtered_df_MAPPED_BALANCED.csv"  # Update with actual CSV file path
OUTPUT_ROOT_DIR = Path("E:\\Sujitha_MISS_DATASET\\2April_FULL_DATASET\\SIEMENS_PPMI_REGISTERED\\")
MNI_PATH = r"E:\Sujitha_MISS_DATASET\2April_FULL_DATASET\MNI152_T2_1mm.nii.gz"
NUM_ROWS = 1  # Set a number to process only a few rows, or None to process all
```

```python
import pandas as pd
import ants
from tqdm import tqdm
```

```python
def register_to_mni(moving_path, fixed_path, output_path):
    """
    Register a T2-weighted MRI to the MNI152 T1 template using ANTsPy's SyN transform.
    """
    moving_path, fixed_path, output_path = Path(moving_path), Path(fixed_path), Path(output_path)
    
    if not moving_path.exists():
        raise FileNotFoundError(f"File not found: {moving_path}")
    if not fixed_path.exists():
        raise FileNotFoundError(f"File not found: {fixed_path}")

    moving = ants.image_read(str(moving_path))
    fixed = ants.image_read(str(fixed_path))

    reg = ants.registration(
        fixed=fixed, moving=moving, type_of_transform="SyN"
    )

    registered = reg["warpedmovout"]

    # Ensure the output directory exists
    output_path.parent.mkdir(parents=True, exist_ok=True)

    ants.image_write(registered, str(output_path))
    print(f"Registered file saved: {output_path}")

def process_csv(csv_path, root_dir, save_root_dir, num_rows=None):
    csv_path, root_dir, save_root_dir = Path(csv_path), Path(root_dir), Path(save_root_dir)
    df = pd.read_csv(csv_path)
    if num_rows:
        df = df.head(num_rows)

    fixed_path = MNI_PATH  # Ensure MNI_PATH is defined
    rows = list(df.iterrows())

    for _, row in tqdm(rows, desc="Processing MRI Files"):
        folder_path = root_dir / Path(row["Path"])
        
        if not folder_path.exists():
            print(f"Warning: Folder not found - {folder_path}")
            continue  # Skip if folder does not exist

        nii_files = list(folder_path.rglob("*.nii.gz"))  # Recursive search in subfolders

        if not nii_files:
            print(f"Warning: No .nii files found in {folder_path}")
            continue

        for nii_file in tqdm(nii_files, desc=f"Processing {row['Path']}", leave=False):
            # Fix the output filename to avoid double .gz
            output_filename = Path(row["Path"]) / f"registered_{nii_file.stem}.nii.gz"  # Use .stem to remove .nii.gz
            output_path = save_root_dir / output_filename

            try:
                register_to_mni(nii_file, fixed_path, output_path)
            except Exception as e:
                print(f"Error processing {nii_file}: {e}")

# Execute the process
process_csv(CSV_FILE, ROOT_DIR, OUTPUT_ROOT_DIR, NUM_ROWS)
```
