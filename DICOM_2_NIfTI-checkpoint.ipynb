```python
# Define root directories
DATASET_ROOTDIR = r"D:\Projects\PROJECT_4_PROFESSOR\Sujitha_Miss\Full_dataset\PPMI_Merged"
OUTPUT_ROOTDIR = r"D:\Projects\PROJECT_4_PROFESSOR\Sujitha_Miss\Full_dataset\PPMI_Merged_NIfTI"
dcm2niix_path = r"D:\Projects\PROJECT_4_PROFESSOR\Sujitha_Miss\dcm2img\MRIcroGL_windows\MRIcroGL\Resources\dcm2niix.exe"
```

```python
import os
import subprocess
from tqdm import tqdm
```

```python
def find_image_folders(root_dir, image_exts=(".dcm")):
    image_folders = []
    for dirpath, _, filenames in tqdm(os.walk(root_dir), desc="Walking Through Directories"):
        if any(f.lower().endswith(image_exts) for f in filenames):
            image_folders.append(dirpath)
    return image_folders

def convert_dicom_to_nifti(input_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)  # Ensure output directory exists
    command = [
        dcm2niix_path,
        "-f", "%f_%p_%t",  # Output filename format
        "-p", "y",  # Preserve patient position
        "-z", "y",  # Compress output NIfTI
        "-ba", "n",  # Do not anonymize
        "-o", output_folder,  # Output directory
        input_folder  # Input DICOM folder
    ]
    subprocess.run(command, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)  # Run command silently

def process_dicom_folders():
    image_folders_list = find_image_folders(DATASET_ROOTDIR)
    print('Total Number of DICOM Folders:', len(image_folders_list))
    
    for dicom_folder in tqdm(image_folders_list, desc="Converting DICOM to NIfTI"):
        relative_path = os.path.relpath(dicom_folder, DATASET_ROOTDIR)
        output_folder = os.path.join(OUTPUT_ROOTDIR, relative_path)  # Maintain directory structure
        convert_dicom_to_nifti(dicom_folder, output_folder)
    
    nifti_folders_list = find_image_folders(OUTPUT_ROOTDIR, image_exts=(".nii", ".nii.gz"))
    print('Total Number of NIfTI Folders:', len(nifti_folders_list))
    print('Total Number of DICOM Data Points:', len(image_folders_list))
    
    print("Conversion Completed!")
```

```python

# Run the processing function
process_dicom_folders()
```
